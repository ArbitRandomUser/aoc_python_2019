```python
def points_from_wire(path):
    x=0;y=1;
    pathpoints = [(0,0)]
    for pathlink in path:
          for i in range(0,int(pathlink[1:])) :
                if pathlink[0]=='R':
                    cp = pathpoints[-1] #current path point
                    np = (cp[x]+1, cp[y])
                    pathpoints.append(np)
                if pathlink[0]=='L':
                    cp = pathpoints[-1] #current path point
                    np = (cp[x]-1, cp[y])
                    pathpoints.append(np)
                if pathlink[0]=='U':
                    cp = pathpoints[-1] #current path point
                    np = (cp[x], cp[y]+1)
                    pathpoints.append(np)
                if pathlink[0]=='D':
                    cp = pathpoints[-1] #current path point
                    np = (cp[x], cp[y]-1)
                    pathpoints.append(np)
    return pathpoints
```

```python
def points_at_distance(dist):
    return [ (i,j) for i in range(-dist,dist+1)\
            for j in range(-dist ,dist+1) \
            if abs(i) + abs(j) == dist]
```
